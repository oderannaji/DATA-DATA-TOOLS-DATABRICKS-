```python
# create dataframe from CSV file.
emails_df = spark.read.csv("/FileStore/tables/assignment_one/emails.csv", header=True, inferSchema=True, quote='"',escape='"', multiLine=True)
emails_df.show(20)
```

```python
row_rdd = emails_df.select("message").rdd

messages_rdd = row_rdd.map(lambda row: row.message.split('\n'))
messages_rdd.take(20)
```

```python
## A1

# First extract the 'From:' field from each email.
from_fields_rdd = messages_rdd.map(lambda email:
    [line.split(': ')[1] for line in email if line.startswith('From: ')][0])

from_fields_rdd.take(100)
```

```python
# 2. Collect all senders into a list
all_senders = from_fields_rdd.collect()
all_senders
```

```python
len(all_senders)
```

```python
from_fields_rdd.takeSample(True, 100)
```

```python
# fetching the unique senders.
unique_senders = set(all_senders)
print(f"unique_senders: {len(unique_senders)}")
```

```python
# mean of the sent emails
sent_emails_mean = len(all_senders) / len(unique_senders)
sent_emails_mean
```

```python
## B2
# Function to filter out non-empty 'To' field
def extract_to_field(email):
    to_lines = [line.split(': ')[1] 
                for line in email
                if line.startswith('To: ')]
    return to_lines[0] if to_lines else None

# function to check missing 'To' field
def check_missing_to(email):
    to_lines = [line for line in email if line.startswith('To: ')]
    return len(to_lines) == 0

# emails with 'To' field
emails_to_rdd = messages_rdd.map(extract_to_field).filter(lambda x: x is not None)

# emails without 'To' field
emails_without_to_rdd = messages_rdd.filter(check_missing_to)

# collect non-existing recipients email
emails_without_to_rdd.take(20)
```
